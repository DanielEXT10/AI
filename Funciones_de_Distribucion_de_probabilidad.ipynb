# Funciones de Distribucion de probabilidades

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Distribucion Uniforme

PFD f(x) =\begin{cases} \frac{1}{b-a} & \text{para } a \leq x \leq b \\ 0 & \text{en otro caso} \end{cases} CFD F(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{x-a}{b-a} & \text{para } a \leq x \leq b \\ 1 & \text{para } x > b \end{cases}

```python
a =1
b = 100
n = 200
data = np.random.uniform(low=a, high=b, size=n)
```

```python
%matplotlib inline
plot = plt.hist(data, bins=20, density=True)
```

### Evaluate for different sample sizes

By increasing the amount of samples histogram starts to look more like a uniform distribution.

```python
sample_sizes = [100, 1000, 10000, 100000]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for i, n in enumerate(sample_sizes):
    data = np.random.uniform(low=a, high=b, size=n)
    axs[i//2, i%2].hist(data, bins=20, density=True)
    axs[i//2, i%2].set_title(f'Uniform Distribution Histogram\nSample Size = {n}')
    axs[i//2, i%2].set_xlabel('Value')
    axs[i//2, i%2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```
