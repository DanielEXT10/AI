```python
results_folder = '/content/drive/MyDrive/Impulse_results'
```

```python
from genericpath import isfile
```

```python
def get_impulse_data(folder):
  files = os.listdir(folder)

  dataframes=[]
  error_files = []

  for file in files:
    full_path = os.path.join(folder,file)
    if os.path.isfile(full_path):

      try:

        f = open(full_path)
        data =json.load(f)
        #Run Number

        fcsn ={'fcsn':data['equipment'][0]['fcsn']}

        try:
          run_info =data['runs'][0]
        except:
          pass

        try:
          temp = data['results']['general']
          t ={key: value.get('value') for key,value in temp.items()}
        except:
          pass



        try:
          environmental = data['results']['environment']
          env ={key: value.get('value') for key, value in environmental.items()}
        except:
          pass

        toolset = data['toolset']
        filtered_toolset = [{'fcsn': d['fcsn'], 'description': d['description'], 'component': d['component']}for d in toolset]
        toolset_dict = {'toolset':toolset}


        tool_dict = {**fcsn,**run_info, **t, **env, **toolset_dict}
        df =pd.DataFrame([tool_dict])
        dataframes.append(df)

      except:
        error_files.append(full_path)
  return dataframes,error_files
```

```python
impulse_df, errors = get_impulse_data(results_folder)
```

```python
Impulse_Data = pd.concat(impulse_df).reset_index()
```
